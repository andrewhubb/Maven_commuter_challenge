```python
def get_forecast_index(test_index, period, granularity):
    """Generate forecast index based on granularity."""
    last_date = test_index[-1]
    if granularity == 'Day':
        return pd.date_range(start=last_date + pd.Timedelta(1, 'D'), periods=period, freq='D')
    elif granularity == 'Week':
        return pd.date_range(start=last_date + pd.Timedelta(7, 'D'), periods=period, freq='W')
    elif granularity == 'Month':
        return pd.date_range(start=last_date + pd.DateOffset(months=1), periods=period, freq='ME')
    elif granularity == 'Quarter':
        return pd.date_range(start=last_date + pd.DateOffset(months=3), periods=period, freq='QE')
    elif granularity == 'Year':
        return pd.date_range(start=last_date + pd.DateOffset(years=1), periods=period, freq='YE')
    raise ValueError(f"Unsupported granularity: {granularity}")

def set_seasonal_order(granularity):
    """Generate forecast index based on granularity."""
    last_date = test_index[-1]
    if granularity == 'Day':
        return best_seasonal_order = (1, 1, 1, 7)
    elif granularity == 'Week':
        return best_seasonal_order = (1, 1, 1, 52)
    elif granularity == 'Month':
        return best_seasonal_order = (1, 1, 1, 12)
    elif granularity == 'Quarter':
        return best_seasonal_order = (1, 1, 1, 4)
    elif granularity == 'Year':
        return best_seasonal_order = (1, 1, 0, 1)
    raise ValueError(f"Unsupported granularity: {granularity}")
def resample_data(data, freq, agg_func='mean'):
    """Resample data based on the specified frequency."""
    if freq is None:  # No resampling for daily data
        return data
    return data.resample(freq).agg(agg_func)

# Frequency mappings
freq_mapping = {
    'Day': None,
    'Week': 'W',
    'Month': 'ME',
    'Quarter': 'QE',
    'Year': 'YE'
}

# User-selected granularity
granularity = 'Month'
period_mapping = {'Day': 365, 'Week': 52, 'Month': 12, 'Quarter': 4, 'Year': 1}
period = period_mapping.get(granularity, 365)

# Best SARIMAX parameters
best_order = (2, 1, 2)
best_seasonal_order = (1, 1, 1, 12)

mta_thousands = create_thousand_dataframe(mta_data)
mta_thousands = mta_data.set_index('Date')

for service in services:
    # Prepare the data for SARIMAX
    if f'{service}: % of Pre-Pandemic' not in mta_thousands:
        print(f"Service {service} data not found. Skipping.")
        continue

    granular_data = resample_data(mta_thousands, freq_mapping[granularity])
    granular_data[f'{service}_Recovery (%)'] = granular_data[f'{service}: % of Pre-Pandemic']
    service_data = granular_data[[f'{service}_Recovery (%)']].dropna().rename(columns={f'{service}_Recovery (%)': 'y'})

    if service_data.empty or len(service_data) <= period:
        print(f"Insufficient data for {service}. Skipping.")
        continue

    # Train/Test split
    train = service_data[:-period]
    test = service_data[-period:]

    try:
        model = SARIMAX(
            train['y'],
            order=best_order,
            seasonal_order=best_seasonal_order,
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        results = model.fit(disp=False)
    except Exception as e:
        print(f"Error fitting SARIMAX for {service}: {e}")
        continue

    # Forecast
    forecast_index = get_forecast_index(test.index, period, granularity)
    forecast_values = results.forecast(steps=period)
    conf_int = results.get_forecast(steps=period).conf_int()

    # Align indices and fill gaps
    unified_index = train.index.union(test.index).union(forecast_index)
    train = train.reindex(unified_index).interpolate(method='time')
    test = test.reindex(unified_index).interpolate(method='time')
    forecast_values = pd.Series(forecast_values, index=forecast_index).reindex(unified_index).interpolate(method='time')
    conf_int = conf_int.reindex(forecast_index).reindex(unified_index, method='nearest').fillna(0)

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(train.index, train['y'], label="Training Data", color='blue')
    plt.plot(test.index, test['y'], label="Test Data", color='green')
    plt.plot(forecast_index, forecast_values.loc[forecast_index], label="Forecast", color='orange')
    plt.fill_between(
        forecast_index,
        conf_int.iloc[:, 0].loc[forecast_index],
        conf_int.iloc[:, 1].loc[forecast_index],
        color='pink',
        alpha=0.3,
        label="Confidence Interval"
    )
    plt.title(f"SARIMAX Forecast for {service} ({granularity} Granularity)")
    plt.xlabel("Date")
    plt.ylabel("Ridership")
    plt.legend()
    plt.show()
```
